```python
from sklearn.datasets import load_breast_cancer
import numpy as np 
import pandas as pd
```

```python
breast = load_breast_cancer()
```

```python
breast.keys()
```

```python
breast.feature_names
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_data
```

```python
breast_labels = breast.target
```

```python
breast_labels.shape
```

```python
breast_labels
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
labels.shape
```

```python
labels
```

```python
final_breast_data = np.concatenate([breast_data, labels],axis=1)
```

```python
final_breast_data.shape
```

```python
final_breast_data
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset.head(10)
```

```python
features = breast.feature_names
```

```python
features
```

```python
breast.keys()
```

```python
features_labels = np.append(features,"label")
```

```python
features_labels
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset.replace({'label': 0},{'label': 'Benign'},inplace=True)
```

```python
breast_dataset['label'] = breast_dataset['label'].replace(1,'Malignant')
```

```python
#breast_dataset['label'].replace(1,'Malignant',inplace=True) #method does not work in python 3.0
```

```python
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = breast_dataset.loc[:,features].values
```

```python
x
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
x.shape
```

```python
np.mean(x), np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
feat_cols
```

```python
normalized_breast = pd.DataFrame(x, columns = feat_cols)
```

```python
normalized_breast.tail()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_breast = PCA(n_components=2)
```

```python
principleComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_df = pd.DataFrame(data = principleComponents_breast, 
                                   columns = ['principal component 1', 'principal component 2'])
```

```python
principal_breast_df.tail()
```

```python
'Explained variability per principal component: {}'.format(pca_breast.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.figure(figsize=(5,5))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principle Component - 1', fontsize=15)
plt.ylabel('Principle Component - 2', fontsize=15)
plt.title('Principle Component Analysis of Breast Cancer Dataset', fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r','g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target 
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'principal component 1']
                , principal_breast_df.loc[indicesToKeep, 'principal component 2'],
                c = color, s = 25)

plt.legend(targets,prop = {'size': 10})
```
